Write a query that returns each genre, with the number of tracks sold in the USA:
in absolute numbers
in percentages.

```python
query1 = """
WITH track_sold_in_USA AS (
SELECT COUNT(InvoiceLineId)
FROM invoice_items  AS it
    JOIN invoices AS i
    ON it.InvoiceId = i.InvoiceID
        JOIN customers AS c
        ON c.CustomerId = i.CustomerID
WHERE c.Country = "USA"
)

SELECT G.Name AS genre,
       COUNT(T.TrackId) AS "number sold", 
       (CAST(COUNT(t.TrackId) as float) / (SELECT * FROM track_sold_in_USA)) * 100 as "percentage sold"
FROM genres as g
    JOIN tracks as t
    ON g.GenreId = t.GenreId
        JOIN invoice_items AS it
        ON t.TrackId = it.TrackId
            JOIN invoices AS i
            ON i.InvoiceID = it.InvoiceID
                JOIN customers AS c
                ON c.CustomerId = i.CustomerId
WHERE c.Country = "USA"
GROUP BY g.GenreId;
"""

cursor = conn.execute(query1)

for i in cursor:
    print(i)
```

Write a query that finds the total dollar amount of sales assigned to each sales support agent within the company. Add any extra attributes for that employee that you find are relevant to the analysis.

```python
query2 = """
SELECT e.EmployeeID,
       e.Title,
       e.HireDate,
       SUM(it.UnitPrice)
FROM employees AS  e
    JOIN customers AS c
    ON c.SupportRepId = e.EmployeeID
        JOIN invoices AS i
        ON i.CustomerId = c.CustomerId
            JOIN invoice_items AS it
            ON i.InvoiceID = it.InvoiceID
GROUP BY e.EmployeeID
"""

cursor = conn.execute(query2)

for i in cursor:
    print(i)
```

Write a query that collates data on purchases from different countries.
Where a country has only one customer, collect them into an "Other" group.
The results should be sorted by the total sales from highest to lowest.
For each country, include:
total number of customers
total value of sales
average value of sales per customer
average order value

```python
query3 = """
SELECT c.Country,
       COUNT(DISTINCT c.CustomerId) AS "Count of customers",
       SUM(it.UnitPrice) AS "Total value of sales",
       SUM(it.UnitPrice) / COUNT(DISTINCT c.CustomerId) AS "Average amount spent by customer",
       SUM(it.UnitPrice) / (COUNT(DISTINCT i.InvoiceId)) AS "Average order value" 
FROM customers AS c    
    JOIN invoices AS i
    ON i.CustomerId = c.CustomerId
        JOIN invoice_items AS it
        ON it.InvoiceId = i.InvoiceId
GROUP BY c.Country
HAVING COUNT(DISTINCT c.CustomerId) > 1

UNION

SELECT "Other",
       COUNT(DISTINCT c.CustomerId) AS "Count of customers",
       SUM(it.UnitPrice) AS "Total value of sales",
       SUM(it.UnitPrice) / COUNT(DISTINCT c.CustomerId) AS "Average amount spent by customer",
       SUM(it.UnitPrice) / (COUNT(DISTINCT i.InvoiceId)) AS "Average order value" 
FROM customers AS c    
    JOIN invoices AS i
    ON i.CustomerId = c.CustomerId
        JOIN invoice_items AS it
        ON it.InvoiceId = i.InvoiceId
WHERE c.Country IN(SELECT Country FROM customers GROUP BY Country HAVING COUNT(*) = 1)
ORDER BY 3 DESC
"""

cursor = conn.execute(query3)

for i in cursor:
    print(i)
```

Which artist is used in the most playlists?

```python
query4 = """
SELECT ar.Name, COUNT(DISTINCT pt.PlaylistId)
FROM artists AS ar
    JOIN albums AS a
    ON a.ArtistId = ar.ArtistId
        JOIN tracks AS t
        ON t.AlbumId = a.AlbumId
            JOIN playlist_track pt
            ON t.TrackId = pt.TrackId
GROUP BY a.ArtistId
ORDER BY 2 DESC
"""

cursor = conn.execute(query4)

for i in cursor:
    print(i)
```

How many tracks have been purchased vs not purchased?

```python
query5 = """
SELECT  COUNT(DISTINCT t.TrackId) AS sold_tracks,
        (SELECT COUNT(*) from tracks) - COUNT(DISTINCT t.TrackId) AS unsold_tracks
FROM tracks AS t
    JOIN invoice_items it
    ON t.TrackId = it.TrackId;
"""

cursor = conn.execute(query5)

for i in cursor:
    print(i)
```
